```python
import plotly.express as px
import plotly.io as pio

import matplotlib.pyplot as plt

import seaborn as sns
import numpy as np
import pandas as pd
import copy
```

```python
df = px.data.gapminder()
df.describe()
```

```python
df.columns.to_list()
```

```python
df.head(n=5)
```

```python
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    color="continent",
    size="pop",
    facet_col="year",
    facet_col_wrap=4,
)
fig.update_yaxes(range=[30, 100])
fig.update_xaxes(range=[0, 50000])
fig.update_layout(height=600)
fig.show()
```

```python
px.bar(
    df[(df["continent"] == "Europe") & (df["year"] == 2007)],
    x="country",
    y="gdpPercap",
    color="continent"
)
```

```python
fig = px.scatter(
    df,
    x="gdpPercap",
    y="lifeExp",
    color="continent",
    size="pop",
    animation_frame="year",
    hover_name="country",
    title="PKB vs Life Expectancy",
)

fig.update_layout(
    title=dict(
        x=0.5,
        xanchor="center"
    ),
    xaxis_title="GDP per capita (USD)",
    yaxis_title="Life expectancy (years)",
    legend_title="Continent",
)
fig.update_yaxes(range=[30, 90])
fig.update_xaxes(range=[20, 45000])

fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
df_2007 = df.query("year == 2007")
```

```python
fig = px.scatter(
    df_2007,
    x="gdpPercap", y="lifeExp",
    size="pop", color="continent",
    hover_name="country",
    log_x=True, size_max=60,
    trendline="ols", trendline_scope="overall",
    title="Wealth vs longevity (2007)",
    labels={"gdpPercap": "GDP per capita (USD)",
            "lifeExp": "Life expectancy (years)"},
)
fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
countries = ["Poland", "Germany", "Japan", "United States", "China", "India"]
df_sub = df[df.country.isin(countries)]

fig = px.line(
    df_sub,
    x="year", y="lifeExp",
    color="country", markers=True,
    title="Life expectancy over time",
    labels={"lifeExp": "Life expectancy (years)"},
)
fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
top15 = df_2007.nlargest(15, "gdpPercap")

fig = px.bar(
    top15,
    x="gdpPercap", y="country",
    color="continent",
    orientation="h",
    title="Top 15 countries by GDP per capita (2007)",
    labels={"gdpPercap": "GDP per capita (USD)"},
)
fig.update_layout(yaxis=dict(categoryorder="total ascending"))
fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
fig = px.box(
    df_2007,
    x="continent", y="lifeExp",
    color="continent",
    points="all",
    hover_name="country",
    title="Life expectancy by continent (2007)",
    category_orders={"continent":
        ["Africa", "Asia", "Americas", "Europe", "Oceania"]},
)
fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
pop_by_cont = df_2007.groupby("continent", as_index=False)["pop"].sum()

fig = px.pie(
    pop_by_cont,
    values="pop", names="continent",
    hole=0.4,
    title="World population by continent (2007)",
)
fig.update_traces(textposition="inside", textinfo="percent+label")
fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
fig = px.choropleth(
    df_2007,
    locations="iso_alpha", color="lifeExp",
    hover_name="country",
    color_continuous_scale="Plasma",
    projection="natural earth",
    title="Life expectancy by country (2007)",
)
fig.update_layout(
    autosize=True,
    width=None,
    height=600,
    margin=dict(l=0, r=0, t=0, b=0),
)
fig.show(config={"responsive": True})
```

```python
fig = px.sunburst(
    df_2007,
    path=["continent", "country"],
    values="pop",
    color="lifeExp",
    color_continuous_scale="Viridis",
    title="World population hierarchy (2007)",
)
fig.update_layout(
    autosize=True,
    width=None,
    height=600,
    margin=dict(l=0, r=0, t=0, b=0),
)
fig.show(config={"responsive": True})
```

```python
fig = px.scatter(
    df,
    x="gdpPercap", y="lifeExp",
    size="pop", color="continent",
    hover_name="country",
    animation_frame="year",
    animation_group="country",
    log_x=True, size_max=60,
    range_x=[100, 100_000], range_y=[25, 90],
    title="Wealth vs longevity over time",
)
fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
fig = px.choropleth(
    df,
    locations="iso_alpha", color="lifeExp",
    hover_name="country",
    animation_frame="year",
    color_continuous_scale="Viridis",
    range_color=[25, 85],
    projection="natural earth",
    title="Life expectancy evolution on world map",
)

fig.update_layout(
    autosize=True,
    width=None,
    height=700,
    margin=dict(l=50, r=50, t=100, b=50),
)
fig.show(config={"responsive": True})
```

```python
# Top 15 countries by GDP per capita, year by year
df_top = (
    df.sort_values(["year", "gdpPercap"], ascending=[True, False])
      .groupby("year")
      .head(15)
      .reset_index(drop=True)
)
fig = px.bar(
    df_top,
    x="gdpPercap", y="country",
    color="continent",
    orientation="h",
    animation_frame="year",
    range_x=[0, df.gdpPercap.max() * 1.05],
    title="Top 15 richest countries per year",
)

fig.update_layout(yaxis={"categoryorder": "total ascending"}, height=600)

fig.show()
```
